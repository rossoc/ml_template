## Data Import

```python
from src.view import Report
import numpy as np
```

```python
def squared(x):
    return x ** 2

def cubed(x):
    return x ** 3
```

```python
report = Report()
figure = report.new_figure()

X = np.linspace(0, 1, 100)

figure.plot({
    r"$x^2$": (X, squared(X)),
    r"$x^3$": (X, cubed(X))
},
    "Functions",
            y_label="Value",
            x_label="X"
)

figure.show()
```
